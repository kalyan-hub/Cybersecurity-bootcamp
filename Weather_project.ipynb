```python
import requests
#import os
from datetime import datetime

api_key = '266c9022bb6e696b530e8879d2f5450e'
location = input("Enter the city name: ")

complete_api_link = "https://api.openweathermap.org/data/2.5/weather?q="+location+"&appid="+api_key
api_link = requests.get(complete_api_link)
api_data = api_link.json()

#create variables to store and display data
temp_city = ((api_data['main']['temp']) - 273.15)
weather_desc = api_data['weather'][0]['description']
hmdt = api_data['main']['humidity']
wind_spd = api_data['wind']['speed']
date_time = datetime.now().strftime("%d %b %Y | %I:%M:%S %p")

print ("-------------------------------------------------------------")
print ("Weather Stats for - {}  || {}".format(location.upper(), date_time))
print ("-------------------------------------------------------------")

print ("Current temperature is: {:.2f} deg C".format(temp_city))
print ("Current weather desc  :",weather_desc)
print ("Current Humidity      :",hmdt, '%')
print ("Current wind speed    :",wind_spd ,'kmph')

report= open('weather2.txt','w')


report.write ("-------------------------------------------------------------"+'\n')
report.write("Weather Stats for - {}  || {}".format(location.upper(), date_time)+ '\n')
report.write("-------------------------------------------------------------" + '\n')

report.write("Current temperature is: {:.2f} deg C".format(temp_city) +'\n')
report.write("Current weather desc  :"+ format(weather_desc) + '\n')
report.write("Current Humidity      :"+ format(hmdt) + '%' + '\n')
report.write("Current wind speed    :"+ format(wind_spd) +'kmph')

report.close()

```
